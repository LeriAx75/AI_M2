```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
```

```python
url = "https://ecobici.cdmx.gob.mx/wp-content/uploads/2025/08/2025-07.csv"
csv_file_name = "data.csv"

try:
    response = requests.get(url, timeout=1200)
    response.raise_for_status()
    print("Descarga completada con éxito")
except requests.exceptions.Timeout as e:
    print(f"Error de timeout durante la descarga: {e}")
    df_raw = pd.DataFrame
except requests.exceptions.RequestException as e:
    print(f"Error durante la descarga: {e}")
    df_raw = pd.DataFrame()
```

```python
with open(csv_file_name, 'wb') as f:
    f.write(response.content)
print(f"Archivo  guardado como {csv_file_name}")

print(f"Leyendo datos desde: {csv_file_name}")
df_ecobici = pd.read_csv(csv_file_name)
df_airbnb = pd.read_csv('listings.csv')
print("Extracción completada con éxito")
```

# Media

```python
media_ecobici = df_ecobici.select_dtypes(include=[np.number]).mean()
media_airbnb = df_airbnb.select_dtypes(include=[np.number]).mean()
```

```python
plt.figure(figsize=(8,4))
media_ecobici.plot(kind='bar', color='skyblue')
plt.title('Medias de ecobici')
plt.ylabel('Medias')
plt.xlabel('Características')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,4))
media_airbnb.plot(kind='bar', color='skyblue')
plt.title('Medias de airbnb')
plt.ylabel('Medias')
plt.xlabel('Características')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
